```python
import pandas as pd
import pmdarima as pm
```

```python
dataL = pd.read_csv('../DataSet/EDvisitfileL.csv', encoding='ISO-8859-1')
dataKa = pd.read_csv('../DataSet/EDvisitfileKa.csv', encoding='ISO-8859-1')
dataKe = pd.read_csv('../DataSet/EDvisitfileKe.csv', encoding='ISO-8859-1')
dataC = pd.read_csv('../DataSet/EDvisitfileC.csv', encoding='ISO-8859-1')
dataT = pd.read_csv('../DataSet/EDvisitfileT.csv', encoding='ISO-8859-1')
dataY = pd.read_csv('../DataSet/EDvisitfileY.csv', encoding='ISO-8859-1')

train_end_L = 3645
train_end_Ka = 3481
train_end_Ke = 3547
train_end_C = 3510
train_end_T = 3645
train_end_Y = 2830

series_data_L = dataL['No']
train_series_data_L = series_data_L.iloc[:train_end_L]

series_data_Ka = dataKa['No']
train_series_data_Ka = series_data_Ka.iloc[:train_end_Ka]

series_data_Ke = dataKe['No']
train_series_data_Ke = series_data_Ke.iloc[:train_end_Ke]

series_data_C = dataC['No']
train_series_data_C = series_data_C.iloc[:train_end_C]

series_data_T = dataT['No']
train_series_data_T = series_data_T.iloc[:train_end_T]

series_data_Y = dataY['No']
train_series_data_Y = series_data_Y.iloc[:train_end_Y]
```

```python
model_L = pm.auto_arima(train_series_data_L,                      
                        seasonal=True,  # 啟用季節性
                        m=7,       # 自動檢測季節性周期
                        stepwise=True,  # 使用階梯式搜索
                        trace=True,     # 打印過程信息
                        error_action='ignore',  # 忽略錯誤
                        suppress_warnings=True)
 
```
