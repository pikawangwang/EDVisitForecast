```python
 
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming you have your time series data in a CSV file
# Load your data
df = pd.read_csv('../DataSet/EDvisitfileLinkou.csv', parse_dates=['date'], index_col='date')

series = df['No']

# 選擇模型進行時序分解
# 可以選擇 'additive' 或 'multiplicative'
result = seasonal_decompose(series, model='multiplicative', period=365)

# Assuming the data column you want to decompose is named 'value'
plt.rcParams.update({'font.size': 12})  # 設定字體大小

# 繪製分解結果
result.observed.plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# 載入您的資料
df = pd.read_csv('../DataSet/EDvisitfileLinkou.csv', parse_dates=['date'], index_col='date')

series = df['No']

# 創建圖表並添加子圖
fig, ax = plt.subplots(3, 1, figsize=(18, 15))  # 調整圖表尺寸以使其更高

# 繪製原始數據
ax[0].plot(series)
ax[0].set_title('ED daily visit')

# 繪製 ACF 圖
sm.graphics.tsa.plot_acf(series, lags=40, ax=ax[1])
ax[1].set_title('Autocorrelation Function (ACF)')

# 繪製 PACF 圖
sm.graphics.tsa.plot_pacf(series, lags=40, ax=ax[2])
ax[2].set_title('Partial Autocorrelation Function (PACF)')

plt.tight_layout()
plt.show()
```
